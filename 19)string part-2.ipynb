 # Accessing Substrings from a string

```python
#concept of Indexing
```

```python
c = 'hello'
print(c)
```

```python
#positive indexing
```

```python
print(c[1])
```

```python
print(c[0])
```

```python
#negative indexing
```

```python
print(c[-1])
```

```python
print(c[-3])
```

```python
#slicing
```

```python
c = 'Hello World'
print(c)
```

```python
print(c[0:5]) # hello-0,1,2,,3,4 count from 0
```

```python
print(c[2:])
```

```python
print(c[4:])
```

```python
print(c[:])
```

```python
print(c[2:6:2])
```

```python
print(c[0:8:3])
```

```python
print(c[0:6:-1]) #does not print +ve index with -ve index
```

```python
print(c[-5:-1:2])
```

```python
print(c[::-1])
```
