```python
#1) Numeric literals
```

```python
a = 0b1010 # binary literals
b = 100 #decimal literals
c = 0o310 #octal literal
d = 0x12c #hexadecimal literals


# float literals 
float_1 = 10.5
float_2 = 1.5e2
float_3 = 1.5e-3

#complex literal
x = 3.14j

print(a,b,c,d)
print(float_1, float_2, float_3)
print(x, x.imag, x.real)
```

```python
#2) string literals
```

```python
string = 'This is Python'
strings = "This is Python"
char = 'C'
multiline_code = """ This is multiline code more than one 
                    line code."""
unicode = u"\U0001F60A". # SMILE CODE different smile different code.
raw_str = r"raw \nstring"

print(string)
print(strings)
print(char)
print(multiline_code)
print(unicode)
print(raw_str)


```
